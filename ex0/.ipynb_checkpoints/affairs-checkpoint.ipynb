```python
# 必要なライブラリを読み込み
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LogisticRegression
from sklearn import tree
from sklearn.cross_validation import train_test_split

# インラインでグラフ表示するおまじない
%matplotlib inline
```

```python
# 結婚している女性に対する、所謂不倫の有無を聞いた調査
# excel ファイルからデータを DataFrame に読み込む
df = pd.read_excel('affairs.xlsx')
```

```python
# データ．先頭から5行を見る
df.head()
```

```python
# 目的変数に「不倫の有無」
Y = df.Had_Affair
```

```python
# 説明変数に「幸せ度」，「年齢」，「結婚年数」，「子供の数」
X = df.loc[:, ['rate_marriage', 'age', 'yrs_married', 'children']]
```

```python
"""
ロジスティック回帰を行います．
"""
log_model = LogisticRegression() # クラスのインスタンス生成
log_model.fit(X, Y)              # 上の目的変数，説明変数を使ってモデル作成
log_model.score(X, Y)            # モデルの精度確認
```

```python
"""
説明変数の寄与を調べる
"""
coeff_df = DataFrame([X.columns, log_model.coef_[0]]).T 
coeff_df
```

```python
"""
決定木分析を行います．
"""
clf = tree.DecisionTreeClassifier(max_depth=3) # クラスのインスタンス生成
clf.fit(X, Y)                                  # 上の目的変数，説明変数を使ってモデル作成
clf.score(X, Y)                                # モデルの精度確認
```

```python
"""
決定木を dot 形式で出力します．
"""
tree.export_graphviz(clf, out_file="tree.dot",
    feature_names=X.columns,
    class_names=Y.name,
    filled=True, rounded=True)
```

```python
"""
dot 形式を png に変換して表示
pydotplus は conda ではなく pip でインストールしました．
Graphviz を別途インストールする必要ありそうです．http://www.graphviz.org/
さらに C:\Program Files (x86)\Graphviz2.38\bin など dot.exe があるフォルダに PATH を通す必要あり
"""
import pydotplus
from IPython.display import Image
dot_data = tree.export_graphviz(clf, out_file=None,
                                feature_names=X.columns,
                                class_names=Y.name,
                                filled=True, rounded=True)
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```
