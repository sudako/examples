## グラフの描画

matplotlib, pandas でも描けるが，seaborn を使うときれいなグラフを簡単に書ける

```python
import numpy as np
from numpy.random import randn
import pandas as pd
import seaborn as sns

%matplotlib inline
```

## 適当なデータを生成
標準正規分布で100個乱数を生成

```python
data1 = randn(100)
data2 = randn(100) + 2 # Off set the mean
```

```python
sns.distplot(data1)
sns.distplot(data2)
```

## 箱ひげ図

```python
sns.boxplot(data=[data1,data2])
```

## バイオリンプロット
箱ひげ図より情報量多い

```python
sns.violinplot(data =[data1,data2])
```
